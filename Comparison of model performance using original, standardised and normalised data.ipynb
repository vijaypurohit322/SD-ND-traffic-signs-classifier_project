```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# Load pickled data

training_file = "traffic-sign-data/train.p"
testing_file = "traffic-sign-data/test.p"

with open(training_file, mode='rb') as f:
    train = pickle.load(f)
with open(testing_file, mode='rb') as f:
    test = pickle.load(f)
    
X_train, y_train = train['features'], train['labels']
X_test, y_test = test['features'], test['labels']
```

```python
# Basic data summary.

n_train = len(X_train)
n_test = len(X_test)
image_shape = X_train[0].shape
n_classes = len(set(y_train))

print("Number of training examples =", n_train)
print("Number of testing examples =", n_test)
print("Image data shape =", image_shape)
print("Number of classes =", n_classes)
```

```python
# Plot four sample images
print('Sample images')
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(X_train[i*1500+1])
```

```python
# Plot a histogram of the count of the number of examples of each sign
# in the test set

plt.hist(y_train, bins=n_classes)
plt.title('Number of examples of each sign in the training set')
plt.xlabel('Sign')
plt.ylabel('Count')
plt.plot()
```

```python
# shuffle
from sklearn.utils import shuffle
X_train, y_train = shuffle(X_train, y_train)
```

```python
X_train_orig = X_train
X_test_orig = X_test
```

```python
# Standardise input (images still in colour)
X_train_std = (X_train - X_train.mean()) / np.std(X_train)
X_test_std = (X_test - X_test.mean()) /np.std(X_test)

# Normalise input (images still in colour)
X_train_norm = (X_train - X_train.mean()) / (np.max(X_train) - np.min(X_train))
X_test_norm = (X_test - X_test.mean()) / (np.max(X_test) - np.min(X_test))
```

```python
def plot_norm_image(image_index):
    plt.subplot(2,2,1)
    plt.imshow(X_train_orig[image_index])
    plt.subplot(2,2,2)
    plt.imshow(X_train_norm[image_index])
    plt.subplot(2,2,3)
    plt.imshow(X_train_std[image_index])
```

```python
plot_norm_image(20)
```

```python
fc_norm_comparison_results = pd.read_csv("logs/fc-keras-norm-comparison.csv")
fc_norm_comparison_results.head()
```

```python
norm_data = fc_norm_comparison_results[fc_norm_comparison_results["preprocessing"] == "norm"]
orig_data = fc_norm_comparison_results[fc_norm_comparison_results["preprocessing"] == "orig"]
stdised_data = fc_norm_comparison_results[fc_norm_comparison_results["preprocessing"] == "std"]
```

```python
stdised_data = stdised_data.rename(lambda x: x - 99)
norm_data = norm_data.rename(lambda x: x + 1)
orig_data = orig_data.rename(lambda x: x - 199)
```

```python
orig_data.plot(x='epoch', title='Original Images')
```

```python
orig_data.plot(y=['acc','val_acc'],x='epoch', title='Original Images')
```

```python
norm_data.plot(x='epoch', title='Normalised Images')
```

```python
stdised_data.plot(x='epoch', title='Standardised Images')
```

```python
acc_combined = pd.DataFrame()
acc_combined['norm'] = norm_data['acc']
acc_combined['stdised'] = stdised_data['acc']
acc_combined['orig'] = orig_data['acc']
```

```python
acc_combined.plot(title="Training Accuracy")
```

```python
acc_combined.plot(y=['norm', 'stdised'], title="Training Accuracy")
```

```python
val_acc_combined = pd.DataFrame()
val_acc_combined['norm'] = norm_data['val_acc']
val_acc_combined['stdised'] = stdised_data['val_acc']
val_acc_combined['orig'] = orig_data['val_acc']
```

```python
val_acc_combined.plot(title="Validation Accuracy")
```
